### load data

```python
import pydicom
from pydicom.data import get_testdata_file
import matplotlib.pyplot as plt
image_extensions = ['dicom'] #only want these file types

def is_dicom_file(file_path):
    try:
        dicom_file = pydicom.dcmread(file_path)
        return True
    except pydicom.errors.InvalidDicomError:
        return False


        
    
def display_images(images):
    for i, image in enumerate(images):
        plt.subplot(3, 5, i + 1)
        plt.imshow(image.pixel_array, cmap=plt.cm.gray)
        plt.axis('off')
        plt.title(f"Image {i + 1}")
    plt.show()

def main():
    data_directory = 'tempdata1/ADNI/'

    # Validate and display DICOM files
    valid_dicom_files = []
    for root, dirs, files in os.walk(data_directory):
        for file in files:
            file_path = os.path.join(root, file)
            if is_dicom_file(file_path):
                valid_dicom_files.append(file_path)

    print(f"Total DICOM files found: {len(valid_dicom_files)}")

    if len(valid_dicom_files) > 0:
        # Display the first 15 images as an example
        images_to_display = []
        for file_path in valid_dicom_files[:15]:
            dicom_file = pydicom.dcmread(file_path)
            images_to_display.append(dicom_file)

        display_images(images_to_display)



# Call the main function when running the notebook
main()
```
